```python
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('mushroom.csv', sep = ';')
```

```python
df.columns
```

```python
for column in df.columns:
    print(df[column].unique())
```

```python
df = df.drop(['veil_type'], axis=1)
```

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
labelencoder_X = LabelEncoder()
labelencoder_y = LabelEncoder()
```

```python
indexes = range(21)
```

```python
for i in indexes:
    X[:,i] = labelencoder_X.fit_transform(X[:,i])
```

```python
y = labelencoder_y.fit_transform(y)
```

```python
df.iloc[:, 1:] = X
df.iloc[:, 0] = y
```

```python
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(df.corr(method='pearson'), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
# Fazendo uma partição para treinamento com o DataFrame com a variável Stalk-root
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75, random_state=42)
```

```python
# Modelo de Árvore de Decisão treinando com a variável Stalk-root
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
y_pred = dt_classifier.predict(X_test)
```

```python
confusion_matrix(y_pred=y_pred, y_true=y_test)
```

É possível ver que o problema da classificação de cogumelos é baseado em regras, logo a Árvore de Decisão não sofreu sobre-ajuste aos dados, mas sim que descobriu tais regras.

Com os experimentos se deram separando a base de dados original entre a parte que possuia todos os atributos presentes e os que não possuiam o stalk-root preenchido, pode-se perceber que é possível propor valores a fim de preencher os valores faltantes deste atributo, dos exemplos que possuiam todos os atributos presentes foi separado 20% para servir de validação e 80% para treinamento de um modelo que pudesse prever qual seria o Stalk-root do exemplo dando as outras características, o modelo treinado obteve 100% de acurácia na base de validação, então foi usado para prever os valores do atributo Stalk-root que não estão presentes na base. 

O resultado obtido pode não estar 100% correto.

A fim de testar o resultado técnicas não baseadas em regras serão usadas e suas acurácias serão avaliadas com os valores preenchidos e sem os valores preenchidos, se houver melhora nos resultados pode-se concluir que mesmo que não esteja 100%, o preenchimento dos dados com o modelo de Árvore é útil.
